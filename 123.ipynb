```python
import matplotlib.pyplot as plt 
import matplotlib.image as img 
import numpy as np
print(np.__version__)
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing import image
from keras.models import load_model

# Data Preprocessing
train_datagen = ImageDataGenerator(rescale=1./255, shear_range=0.2, zoom_range=0.2, horizontal_flip=True)
test_datagen = ImageDataGenerator(rescale=1./255)
training_set = train_datagen.flow_from_directory('E:/devanagiri-character-recognition-master/dataset/training_set', color_mode='grayscale',target_size=(32, 32), batch_size=32, class_mode='categorical')
test_set = test_datagen.flow_from_directory('E:/devanagiri-character-recognition-master/dataset/test_set', color_mode='grayscale', target_size=(32, 32), batch_size=32, class_mode='categorical')


# Initialize the CNN
classifier = Sequential()

# Add Convolution2D layer 
classifier.add(Conv2D(32, (3, 3), input_shape = (32, 32, 1), activation='relu'))

# Pooling
classifier.add(MaxPooling2D(pool_size=(2, 2)))

classifier.add(Conv2D(32, (3, 3), activation='relu'))
classifier.add(MaxPooling2D(pool_size=(2, 2)))

classifier.add(Conv2D(32, (3, 3), activation='relu'))
classifier.add(MaxPooling2D(pool_size=(2, 2)))

# Flattening
classifier.add(Flatten())

# Fully connected network
classifier.add(Dense(units=128, activation='relu'))
classifier.add(Dense(units=128, activation='relu'))
classifier.add(Dense(units=64, activation='relu'))
classifier.add(Dense(units=46, activation='softmax'))
classifier.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Fitting the CNN
classifier.fit_generator(training_set, steps_per_epoch=2588, epochs=1, validation_data=test_set, validation_steps=288)

# Save the model
classifier.save('E:/devanagiri-character-recognition-master/devanagiri.h5')


# Load the model
#classifier = load_model('devagiri.h5')

# Single prediction
test_image = image.load_img('E:/devanagiri-character-recognition-master/dataset/single_prediction/size.png', grayscale=True, target_size = (32, 32))
test_image = image.img_to_array(test_image)
test_image = np.expand_dims(test_image, axis = 0)
result = classifier.predict(test_image)

  
# reading png image file 
im = img.imread('E:/devanagiri-character-recognition-master/dataset/single_prediction/size.png') 
  
# show image 
plt.imshow(im) 
from PIL import Image
img = Image.open("E:/devanagiri-character-recognition-master/dataset/single_prediction/size.png")
img.show()

#training_set.class_indices
if result[0][0] == 1:
    print('ka')
elif result[0][1] == 1:
    print('kha')
elif result[0][2] == 1:
    print('ga')
elif result[0][3] == 1:
    print('gha')
elif result[0][4] == 1:
    print('kna')
elif result[0][5] == 1:
    print('cha')
elif result[0][6] == 1:
    print('chha')
elif result[0][7] == 1:
    print('ja')
elif result[0][8] == 1:
    print('jha')
elif result[0][9] == 1:
    print('yna')
elif result[0][10] == 1:
    print('taamatar')
elif result[0][11] == 1:
    print('thaa')
elif result[0][12] == 1:
    print('daa')
elif result[0][13] == 1:
    print('dhaa')
elif result[0][14] == 1:
    print('adna')
elif result[0][15] == 1:
    print('tabala')
elif result[0][16] == 1:
    print('tha')
elif result[0][17] == 1:
    print('da')
elif result[0][18] == 1:
    print('dha')
elif result[0][19] == 1:
    print('na')
elif result[0][20] == 1:
    print('pa')
elif result[0][21] == 1:
    print('pha')
elif result[0][22] == 1:
    print('ba')
elif result[0][23] == 1:
    print('bha')
elif result[0][24] == 1:
    print('ma')
elif result[0][25] == 1:
    print('yaw')
elif result[0][26] == 1:
    print('ra')
elif result[0][27] == 1:
    print('la')
elif result[0][28] == 1:
    print('waw')
elif result[0][29] == 1:
    print('motosaw')
elif result[0][30] == 1:
    print('petchiryakha')
elif result[0][31] == 1:
    print('patalosaw')
elif result[0][32] == 1:
    print('ha')
elif result[0][33] == 1:
    print('chhya')
elif result[0][34] == 1:
    print('tra')
elif result[0][35] == 1:
    print('gya')
elif result[0][36] == 1:
    print('0')
elif result[0][37] == 1:
    print('1')
elif result[0][38] == 1:
    print('2')
elif result[0][39] == 1:
    print('3')
elif result[0][40] == 1:
    print('4')
elif result[0][41] == 1:
    print('5')
elif result[0][42] == 1:
    print('6')
elif result[0][43] == 1:
    print('7')
elif result[0][44] == 1:
    print('8')
else:
    print('9')
```
